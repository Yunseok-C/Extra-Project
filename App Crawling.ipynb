### appfollow

```python
browser.get('https://appfollow.io/ko/rankings/iphone/kr/business#2020-12-30')
```

```python
soup = BeautifulSoup(browser.page_source)
payment = ['유료','무료']
appdata = []
temp = soup.find_all('div', {'class':'rankings__table-block'})[0]
for l1 in temp:
    temp2 = l1.find_all('div', {'class':'rankings__app-info'})
    for l2 in temp2 :
        tlist = []
        tlist.append(l2.find('a').text)
        tlist.append(soup.find('div',{'class':'text','role':'alert'}).find('span').text)
        tlist.append(int(l2.find('div',{'class':'rating-stars__avg active'}).attrs['style'].split(' ')[1].rstrip('%;'))/20)
        tlist.append(l2.find('div',{'class':'rating-stars__count'}).text.lstrip('(').rstrip(')'))
        tlist.append(l2.find('div',{'class':'rankings__app-price'}).text)
        appdata.append(tlist)
temp = soup.find_all('div', {'class':'rankings__table-block'})[1]
for l1 in temp:
    temp2 = l1.find_all('div', {'class':'rankings__app-info'})
    for l2 in temp2 :
        tlist = []
        tlist.append(l2.find('a').text)
        tlist.append(soup.find('div',{'class':'text','role':'alert'}).find('span').text)
        tlist.append(int(l2.find('div',{'class':'rating-stars__avg active'}).attrs['style'].split(' ')[1].rstrip('%;'))/20)
        tlist.append(l2.find('div',{'class':'rating-stars__count'}).text.lstrip('(').rstrip(')'))
        tlist.append(l2.find('div',{'class':'rankings__app-price'}).text)
        appdata.append(tlist)
```

```python
table = pd.DataFrame(appdata, columns= ['이름','카테고리','별점','평가수','가격'])
table.to_excel(excel_writer = r'C:\Users\com\Desktop\US_iphone_{}.xls'.format(appdata[0][1]))
```

### App Annie

```python
browser.get('https://www.appannie.com/apps/ios/top-chart?country=IN&category=6014&device=iphone&date=2020-12-30&feed=Free&rank_sorting_type=rank&page_number=0&page_size=100&metrics=grossing_rank,category,all_avg,all_count,last_avg,last_count,first_release_date,last_updated_date,est_download,est_revenue,wau&order_type=desc&order_by=free_rank')
```

```python
browser.switch_to_frame(browser.find_element_by_xpath('//*[@id="__next"]/div[2]/div/div[2]/div[2]/div/div/iframe'))
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
applist = []
i = 0
for app in soup.find_all('tr',{'class':'main-row table-row'}):
    #print(i)
    title = app.select('a.app-link>span')[0].text
    category = app.select('td.text')[0].text
    payment = '무료'
    #payment = app.select('td>span>span')[0].get_text()
    if app.select('div.rating-value') == []:
        star_rating = 'N/A'
    else:
        star_rating = app.select('div.rating-value')[0].text
    rating = app.select('td.number>span')[1].text
    applist.append([title, category, payment,star_rating,rating])
    i +=1
    if i == 100:
        break
```

```python
applist
```

```python
pd.DataFrame(applist, columns = ['제목','카테고리','가격','평점','평가수']).to_excel(excel_writer = r'C:\Users\com\Desktop\US_ios_free.xls')
```

```python
data = []
wait = WebDriverWait(browser, 10)
for i in range(14):
    wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "n_1jv8115-o_O-Content_1xn1r0f-o_O-c0f2346_4s8r2l-o_O-text_7y41qf-o_O-item_9xo99e")))
    time.sleep(1)
    browser.find_element_by_xpath('//*[@id="react-app"]/div[2]/div[3]/div[3]/a[{}]'.format(i+1)).click()
    wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "container_1jii25e")))
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    title = soup.find('h1').text
    price = browser.find_element_by_xpath('//*[@id="react-app"]/section/div/div[2]/div/div[1]/div[2]/span[2]').text
    star = soup.find('div',{'class': 'star-ratings'})['title']
    comments = soup.find('h4',{'style':True}).text.split(' ')[0]
    data.append([title, price, star, comments])
    browser.back()
#data
```

```python
table = pd.DataFrame(data,columns=['Title','Price','Rating','Comments'])
table.to_excel(excel_writer = r'C:\Users\com\Desktop\US_ios_Income.xlsx')
```
